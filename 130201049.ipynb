#### Problem 1: Ülkelerin COVID-19 tanısı konulmuş vaka sayısı büyükten küçüğe 

```python
query = """
        SELECT 
            Country_Region,
            SUM(Confirmed) as Confirmed 
        FROM covid19 
        GROUP BY Country_Region 
        ORDER BY Confirmed DESC 
        LIMIT 20
        """
df = pd.read_sql_query(query,connection)
df
```

#### Problem 2: COVID-19 tanısı konulmuş vaka sayısı en az 10000 olan ülkelerde ölüm oranları büyükten küçüğe

```python
query = """
        SELECT 
            Country_Region, 
            SUM(Confirmed) as Confirmed, 
            SUM(Deaths) as Deaths, 
            SUM(Deaths)*100.0/SUM(Confirmed) as percent
        FROM covid19 
        GROUP BY Country_Region 
        HAVING SUM(Confirmed) > 10000
        ORDER BY percent DESC 
        LIMIT 20
        """
df = pd.read_sql_query(query,connection)
pd.set_option('precision',2)
df
```

#### Problem 2: COVID-19 tanısı konulmuş vaka sayısı en az 10000 olan ülkelerde iyileşme oranları büyükten küçüğe

```python
query = """
        SELECT
            Country_Region, 
            SUM(Confirmed) as Confirmed,
            SUM(Recovered) as Recovered,
            SUM(Recovered)*100.0/SUM(Confirmed) as percent
        FROM covid19
        GROUP BY Country_Region
        HAVING SUM(Confirmed) > 10000
        ORDER BY percent DESC
        LIMIT 20
        """
df = pd.read_sql_query(query,connection)
pd.set_option('precision',2)
df
```

#### Problem 13: 1976 ve 1988 yıllarındaki olimpiyat oyunlarında Doğu ve Batı Almanya'nın aldığı madalyaların cinsiyete göre gösterimi

```python
query = """
        SELECT 
            Gender, 
            SUM(CASE WHEN Country='East Germany' THEN 1 ELSE 0 END) as 'East Germany',
            SUM(CASE WHEN Country='West Germany' THEN 1 ELSE 0 END) as 'West Germany',
            COUNT(Medal)
        FROM Clean_Summer
        WHERE 
            Country IN ('East Germany', 'West Germany') AND
            Year IN (1976, 1988)
        GROUP BY Gender"""
df = pd.read_sql_query(query,connection)
df
```

#### Problem 14: Basketbolda madalya kazanan ülkeler ve hangi madalyayı kaç defa kazandıklarının bilgisi

```python
#Kadroda 12 oyuncu yer aldığı için 12'ye bölüyorum. 
query = """
        SELECT
            Country,
            ROUND((SUM(CASE WHEN Medal='Gold' THEN 1 ELSE 0 END)/12),0) as Gold,
            ROUND((SUM(CASE WHEN Medal='Silver' THEN 1 ELSE 0 END)/12.0),0) as Silver,
            ROUND((SUM(CASE WHEN Medal='Bronze' THEN 1 ELSE 0 END)/12),0) as Bronze,
            ROUND(COUNT(Medal)/12.0,0) as 'Total Medal'
        FROM Clean_Summer
        WHERE Sport='Basketball' AND Gender='Men'
        GROUP BY Country
        ORDER BY Gold DESC, Silver DESC, Bronze DESC"""
df = pd.read_sql_query(query,connection)
pd.set_option('precision',0)
df
```
